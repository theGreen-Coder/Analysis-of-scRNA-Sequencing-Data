```python
import pandas as pd
import ast
```

```python
regulon = pd.read_csv("./regSignificant.csv")
```

```python
regulon
```

## Filter by Context

```python
allContexts = [
    "weight>75.0",
    "weight>90.0",
    "top50",
    "top5perTarget",
    "top10perTarget"
]
selectedFilter = allContexts[0]
```

```python
fRegulons = regulon[regulon['Context'].str.contains(selectedFilter)]
```

```python
fRegulons
```

## Build Network

```python
## Get All Genes from dataset
allRegulons = fRegulons["TF"].to_list()
allRegulons = list(dict.fromkeys(allRegulons))
allRegulons
```

```python
network = pd.DataFrame(columns=['source', 'target', 'value'])
network
```

```python
for regulonGene in allRegulons:
    regulonDF = fRegulons[fRegulons["TF"] == regulonGene]
    regulonDF = regulonDF.reset_index()
    i = 0
    for i in regulonDF.index:
        transcriptionFactor = regulonDF.iloc[i]["TF"]
        targetGenesStr = regulonDF.iloc[i]["TargetGenes"]
        targetGenes = ast.literal_eval(targetGenesStr)

        for tupleGene in targetGenes:
            if transcriptionFactor != tupleGene[0]:
                network.loc[len(network.index)] = [transcriptionFactor, tupleGene[0], tupleGene[1]]

network = network.drop_duplicates()
```

```python
network
```

```python
network = network[(network["source"] == "BHLHE22")]
network
```

## Plot to Cytoscape

```python
network.to_csv(f"./results/resultsBHLHE22{selectedFilter.upper()}.csv", index=False)
```
